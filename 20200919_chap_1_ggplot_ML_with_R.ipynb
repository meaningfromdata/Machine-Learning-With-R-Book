```
library(tidyverse)
```

```
data(iris)
```

```
head(iris)
```

```
myPlot <- ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width)) +
  geom_point() + 
  theme_bw()
```

```
myPlot
```

```
# adding contour plot and loess line through point
myPlot + 
  geom_density_2d() + 
  geom_smooth()
```

```
# chaning ordering so contour plot is lowest layer, then points, then loess line
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width)) +
  geom_density_2d() +
  geom_point() + 
  geom_smooth() +
  theme_bw()
```

```
# adding shape of point by species
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, shape = Species)) +
  geom_point() + 
  theme_bw()
```

```
# coloring points by species
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, col = Species)) +
  geom_point() + 
  theme_bw()
```

```
# adding faceting by species
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, col = Species)) +
  facet_wrap(~ Species) +
  geom_point() + 
  theme_bw()
```
