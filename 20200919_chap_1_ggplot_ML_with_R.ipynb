```
myPlot <- ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width)) +
  geom_point() + 
  theme_bw()
```

```
myPlot
```

```
myPlot + 
  geom_density_2d() +
  geom_smooth()
```
